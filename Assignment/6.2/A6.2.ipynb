```python
df=df.fillna({'language':'No data','country': 'No data','color': 'No data'})
```

```python
df=df.dropna(subset=['duration','title_year','budget','gross', 'num_critic_for_reviews'])
```

```python
# Downsizing dataset
df['title_year']=df['title_year'].astype('int16')
df['duration']=df['duration'].astype('int16')
df['num_critic_for_reviews']=df['num_critic_for_reviews'].astype('int16')
df['director_facebook_likes']=df['director_facebook_likes'].astype('int16')
df['budget']=df['budget'].astype('int64')
df['gross']=df['gross'].astype('int64')
df['imdb_score']=df['imdb_score'].astype('int16')
```

```python
# Check for duplicate

for col in df.columns.tolist():
      weird = (df[[col]].applymap(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
      if len (df[weird]) > 0:
        print (col)
        
```

```python
df['actor_1_name'] = df['actor_1_name'].astype('str')
df['actor_2_name'] = df['actor_2_name'].astype('str')
df['actor_3_name'] = df['actor_3_name'].astype('str')
df['content_rating'] = df['content_rating'].astype('str')
df['plot_keywords'] = df['plot_keywords'].astype('str')
```

```python
for col in df.columns.tolist():
      weird = (df[[col]].applymap(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
      if len (df[weird]) > 0:
        print (col)
```

## Correlation heatmap

```python
# Create a correlation heatmap using matplotlib
# Save figure
plt.matshow(df.corr())
plt.savefig("Netcorr.png") 
```

```python
#current dir
cwd = os.getcwd()
cwd
```

```python
# Add labels, a legend, and change the size of the heatmap

f = plt.figure(figsize=(8, 8)) # figure size 
plt.matshow(df.corr(), fignum=f.number) # type of plot
plt.xticks(range(df.shape[1]), df.columns, fontsize=12, rotation=80) # x axis labels
plt.yticks(range(df.shape[1]), df.columns, fontsize=14) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=14) # add font size
plt.title('Correlation Matrix', fontsize=14) # add title
```

Looking at the heatmap, it seems like the factors don't have a strong relationship. some of them have a moderate relationship such as when gross income of a movie is high, the number of critic for reviews is high. Also, new movies have more number of critic for reviews than the old movies as well.

## Categorical Plots

```python
# Use a histogram to visualize the distribution of the variables. 

sns.histplot(df['duration'], bins = 20, kde = True)
```

```python
df.loc[df['duration'] < 80, 'movie_time'] = 'short movie'
df.loc[(df['duration'] >= 70) & (df['duration'] < 180), 'movie_time'] = 'average movie'
df.loc[df['duration'] >= 180, 'movie_time'] = 'long movie'
```

```python
df['movie_time'].value_counts(dropna = False)
```

```python
df.columns
```

```python
# Create a categorical plot in seaborn using the price categories created above

sns.set(style="ticks")
g = sns.catplot(x="imdb_score", y="duration", hue="movie_time", data=df)
```

```python
# fig = g.get_figure()
g.savefig("Netcplot.png") 
```

Based on the categorical plot, it's easy to "short movie" falls between imdb score of 6 and 7 and "long movies" have scores 5 and over. and the rest of the "average movies" fall under all imdb score.
It could mean if you want to have a movie with a imdb score 5 and over, it will be good to choose either a short movie or a long movie.
